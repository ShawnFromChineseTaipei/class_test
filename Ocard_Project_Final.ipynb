```python
import pandas as pd
df = pd.read_csv("Ocard_point.csv")
del df['Unnamed: 0']
```

```python
df.head(20)
```

```python
df['交易時間'] = pd.to_datetime(df['交易時間'])
df['交易時間_年'] = df['交易時間'].dt.year
df['交易時間_月'] = df['交易時間'].dt.month
df['交易時間_日'] = df['交易時間'].dt.day
```

```python
# 1. 將資料按照依照相同的用戶編號由上而下作排序。
df.sort_values(by=['用戶編號', '交易時間'], inplace=True)

# 2. 計算相同的用戶編號共有多少rows，並將這個數值產生新特徵，命名為Consumtion_times
df['Consumtion_times'] = df.groupby('用戶編號')['用戶編號'].transform('count')

# 3. 將相同用戶編號的rows中的交易時間，兩兩做相減，得出新的特徵Date_diff，以日為單位。
df['Date_diff'] = df.groupby('用戶編號')['交易時間'].diff().dt.days

# 4. 將相同用戶編號的rows中的Date_diff，全部相加除以Consumtion_times，得到新特徵，命名為Average_Date_diff
df['Average_Date_diff'] = df.groupby('用戶編號')['Date_diff'].transform('mean')

# 移除 'Average_Date_diff' 中的 NaN 值
df.dropna(subset=['Average_Date_diff'], inplace=True)

# 移除 'Average_Date_diff' 中的 0 值
df = df[df['Average_Date_diff'] != 0]
```

```python
df.describe()
```

```python
df2 = pd.read_csv('Ocard dataset.csv')
```

```python
# 將 df2 中的 'age' 和 'gender' 加入 df 中
df = df.merge(df2, left_on='用戶編號', right_on='ID', how='left')

# 然後你可能不需要 'ID' 這個欄位，因為它和 '用戶編號' 是一樣的
df = df.drop(columns='ID')

# 再次確認數據
df.info()
```

```python
df.head(50)
```

```python


# 刪除模型中不需要的特徵
df = df.drop(columns=['動作', '兌換品項', '交易序號', '禮物券名稱', 'Level', '加入時間', 'Spending amount', '核銷時間',
                      '發放店家','交易時間_年','交易時間_月','交易時間_日','Date_diff'])
```

```python
df['核銷店家'] = df['核銷店家'].astype(str)
# 處理 '核銷店家' 特徵，刪除 "-" 之後的文字
df['核銷店家'] = df['核銷店家'].apply(lambda x: x.split('-')[0])
```

```python
df.head(50)
```

```python
df = df.drop(columns=['分店','交易時間','消費店名' ])
```

```python
df = pd.get_dummies(df, columns=['gender'])
unique_stores = df['核銷店家'].unique()
print(unique_stores)
```

```python
stores_to_keep = ['洋城義大利餐廳', '米塔義式廚房', '洋朵庭園餐廳', '千荷田蔬食百匯涮涮鍋', 
                  '千荷田火鍋百匯涮涮鍋', '湯師傅鍋物專賣', '大河屋', '日本橋浜町食事処', 
                  'Mita CAFE Bistro', '米塔炙燒牛排專門店', '泰舍']
df = pd.get_dummies(df, columns=['Redeem_weekday'])

df = df[df['核銷店家'].isin(stores_to_keep)]
```

```python
import numpy as np
def calculate_customer_level(days):
    if days < 30:
        return 0
    elif 30<days < 90:
        return 1
    elif 90<days < 180:
        return 2
    else:
        return np.nan

df['customer_level'] = df['Average_Date_diff'].apply(calculate_customer_level)
df = df.dropna(subset=['customer_level'])
df.head(50)
df.info()
```

```python
df = df.dropna()
print(df.isnull().sum())
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier


X = df.drop(['用戶編號', '核銷店家', 'customer_level','Redeem_date','Average_Date_diff'], axis=1)
y = df['customer_level']

# 因為數據太多了QAQ，我只抽樣10000筆數據
X_sample = X.sample(n=10000, random_state=1)
y_sample = y[X_sample.index]


X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, random_state=1)


sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# SVM
svc = SVC()
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
print('SVC Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_svc))
print('SVC Accuracy: ', accuracy_score(y_test, y_pred_svc))

# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
print('Naive Bayes Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_nb))
print('Naive Bayes Accuracy: ', accuracy_score(y_test, y_pred_nb))

# XGBoost
xgb = XGBClassifier(use_label_encoder=False)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
print('XGBoost Confusion Matrix:')
print(confusion_matrix(y_test, y_pred_xgb))
print('XGBoost Accuracy: ', accuracy_score(y_test, y_pred_xgb))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix

# 定義一個函數來顯示模型的性能指標
def print_metrics(y_test, y_pred):
    print('Accuracy: ', accuracy_score(y_test, y_pred))
    print('Precision: ', precision_score(y_test, y_pred, average='weighted'))
    print('Recall: ', recall_score(y_test, y_pred, average='weighted'))
    print('F1 Score: ', f1_score(y_test, y_pred, average='weighted'))

# SVM
print('SVM Performance:')
print_metrics(y_test, y_pred_svc)
plot_confusion_matrix(svc, X_test, y_test)
plt.show()

# Naive Bayes
print('Naive Bayes Performance:')
print_metrics(y_test, y_pred_nb)
plot_confusion_matrix(nb, X_test, y_test)
plt.show()

# XGBoost
print('XGBoost Performance:')
print_metrics(y_test, y_pred_xgb)
plot_confusion_matrix(xgb, X_test, y_test)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
X, y = X_sample, y_sample
title = "Learning Curves (SVM)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
estimator = SVC()
plot_learning_curve(estimator, title, X, y, (0.7, 1.01), cv=cv, n_jobs=4)

title = "Learning Curves (Naive Bayes)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
estimator = GaussianNB()
plot_learning_curve(estimator, title, X, y, (0.7, 1.01), cv=cv, n_jobs=4)

# XGBoost Learning Curve
title = "Learning Curves (XGBoost)"
cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
estimator = XGBClassifier(use_label_encoder=False)
plot_learning_curve(estimator, title, X, y, (0.7, 1.01), cv=cv, n_jobs=4)

plt.show()
```
