```python
import pandas as pd
import matplotlib.pyplot as plt


df1 = pd.read_csv('dogg.csv')
df1.head(5)
#animal_id(動物的流水編號)、animal_subid(動物的區域編號)、animal_area_pkid(動物所屬縣市代碼)、
#animal_shelter_pkid(動物所屬收容所代碼)、animal_place(動物的實際所在地)、animal_kind(動物的類型)、
#animal_Variety(動物品種)、animal_sex(動物性別)、animal_bodytype(動物體型)、animal_colour(動物毛色)、
#animal_age(動物年紀)、animal_sterilization(是否絕育)、animal_bacterin(是否施打狂犬病疫苗)、animal_foundplace(動物尋獲地)、
#animal_title(動物網頁標題)、animal_status(動物狀態)、animal_remark(資料備註)、animal_caption(其他說明)、animal_opendate(開放認養時間(起))、
#animal_closeddate(開放認養時間(迄))、animal_update(動物資料異動時間)、animal_createtime(動物資料建立時間)、shelter_name(動物所屬收容所名稱)、
#album_file(圖片名稱)、album_update(異動時間)、cDate(資料更新時間)、shelter_address(地址)、shelter_tel(連絡電話)
```

```python
df1 = df1.drop(["animal_subid","animal_id","animal_remark","animal_caption","animal_title","animal_update","cDate","album_file","shelter_tel","album_update","shelter_address","shelter_name","animal_place","animal_foundplace"],axis=1)
df1
```

```python
#查看欄位資料屬性
df1.dtypes
#並刪除missing value
df1.dropna()
```

```python
#為了方便後面分析 將動物品種以及年紀更改為0或1，狗=0,貓=1 ；幼小=0,成人=1 以此類推
relation_map = {'狗':0,'貓':1,'ADULT':1,'CHILD':0,"T":0,"F":1,"M":0}
df1["Animal_kind"] = df1["animal_kind"].map(relation_map)
df1["Animal_age"] = df1["animal_age"].map(relation_map)
df1["Animal_sex"] = df1["animal_sex"].map(relation_map)
df1["Animal_sterilization"] = df1["animal_sterilization"].map(relation_map)
#並刪除原來的欄位，只保留有數值的欄位
df2 = df1.drop(["animal_kind","animal_age","animal_sex","animal_sterilization"],axis=1)
df2=df2.dropna()
```
