```python
import pandas as pd
import datetime as dt

# 載入廣告資料
df = pd.read_csv('Ocard dataset.csv')
df
```

```python
df.info()
```

```python
missing_values = df[['Level', 'age', 'gender']].isna().sum()
print("缺失值数量：")
print(missing_values)
```

```python
df.dropna(inplace=True)
missing_values = df.isna().sum()
print("缺失值数量：")
print(missing_values)
```

## 新增欄位（加入時間相關）
'Join_weekday', 'Join_year', 'Join_month', 'Join_day', 'Join_date'

```python
df['加入時間'] = pd.to_datetime(df['加入時間'])


df['Join_weekday'] = df['加入時間'].dt.strftime('%A')


df['Join_date'] = df['加入時間'].dt.day

df['Join_year'] = df['加入時間'].dt.year
df['Join_month'] = df['加入時間'].dt.month

df['Join_year'] = df['加入時間'].dt.year
df['Join_month'] = df['加入時間'].dt.month
df['Join_day'] = df['加入時間'].dt.day


df['Join_date'] = df['Join_year'].astype(str) + '-' + df['Join_month'].astype(str) + '-' + df['Join_day'].astype(str)


print(df[['ID', '加入時間', 'Join_weekday', 'Join_year', 'Join_month', 'Join_day', 'Join_date']].head(10))
```

## 新增欄位（核銷時間相關）

```python
df['核銷時間'] = pd.to_datetime(df['核銷時間'])


df['Reimburse_weekday'] = df['核銷時間'].dt.strftime('%A')


df['Reimburse_date'] = df['核銷時間'].dt.day

df['Reimburse_year'] = df['核銷時間'].dt.year
df['Reimburse_month'] = df['核銷時間'].dt.month

df['Reimburse_year'] = df['核銷時間'].dt.year
df['Reimburse_month'] = df['核銷時間'].dt.month
df['Reimburse_day'] = df['核銷時間'].dt.day


df['Reimburse_date'] = df['Reimburse_year'].astype(str) + '-' + df['Reimburse_month'].astype(str) + '-' + df['Reimburse_day'].astype(str)


print(df[['ID', '核銷時間', 'Reimburse_weekday', 'Reimburse_year', 'Reimburse_month', 'Reimburse_day', 'Reimburse_date']].head(10))
```
