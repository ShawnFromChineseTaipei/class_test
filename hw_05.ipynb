```python
import pandas as pd
df = pd.read_csv("Salary_Data.csv")
df
```

```python
import numpy as np
x = np.array(df["YearsExperience"])
y = np.array(df["Salary"])
x = x.reshape(-1,1)
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
scaler = StandardScaler()
scaler.fit(x)
std_x = scaler.transform(x)



print(std_x)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


train_x, test_x, train_y, test_y  = train_test_split(std_x, y, test_size=0.2, random_state=10)

regression = LinearRegression()
regression.fit(train_x, train_y)
y_p = regression.predict(test_x)
Y_pred = regression.predict(std_x)
y_pred = Y_pred.reshape(-1,1)
orginal_x = scaler.inverse_transform(y_pred)
print(orginal_x)

print(f"evaluation MSE: {mean_squared_error(test_y, y_p)}")
print(f"x的真實值為{y_p},預測結果為{Y_pred}")

plt.scatter(std_x[:, 0], y, s = 3)
plt.plot(std_x,Y_pred, color = 'red')
plt.show()
```
